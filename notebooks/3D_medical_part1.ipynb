```python
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
t1_image_path = "/Users/a200005921/PROJECTS_DEV/LEARNING/MANNING_3D_MEDICAL/data/small/t1/IXI130-HH-1528-T1_fcm.nii.gz"
t2_image_path = "/Users/a200005921/PROJECTS_DEV/LEARNING/MANNING_3D_MEDICAL/data/small/t2/IXI130-HH-1528-T2_reg_fcm.nii.gz"
```

```python
t1_image = nib.load(t1_image_path)
t1_image_data = t1_image.get_fdata()

t2_image = nib.load(t2_image_path)
t2_image_data = t2_image.get_fdata()
```

```python
def show_slices(slices):
    """ Function to display row of image slices """
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
       axes[i].imshow(slice.T, cmap="gray", origin="lower")
```

```python
t1_slice_0 = t1_image_data[26, :, :]
t1_slice_1 = t1_image_data[:, 30, :]
t1_slice_2 = t1_image_data[:, :, 16]
```

```python
t2_image = nib.load(t2_image_path)
t2_image_data = t2_image.get_fdata()
```

```python
t2_slice_0 = t2_image_data[26, :, :]
t2_slice_1 = t2_image_data[:, 30, :]
t2_slice_2 = t2_image_data[:, :, 16]
```

```python
show_slices([t1_slice_0, t2_slice_0])
plt.suptitle("Center slices for T1 and T2 image") 
```

```python
show_slices([t1_slice_1, t2_slice_1])
plt.suptitle("Center slices for T1 and T2 image") 
```

```python
show_slices([t1_slice_2, t2_slice_2])
plt.suptitle("Center slices for T1 and T2 image") 
```
